```python
# 1. Загрузить набор данных из репозитория. Свой выбор данных  согласовать с преподавателем.
import pandas as pd

df = pd.read_csv("./data/global_cancer_patients_2015_2024.csv")

print(df.info())

df = df.drop_duplicates()

df.head()
```

```python
# 5. Отобрать и определить признаки, которые будут использоваться для построения моделей.

important_feat = [
    "Age",
    "Gender",
    "Country_Region",
    "Genetic_Risk",
    "Air_Pollution",
    "Alcohol_Use",
    "Smoking",
    "Obesity_Level",
    "Cancer_Type"
    "Cancer_Stage",
]

numeric_columns = [
    "Age",
    "Genetic_Risk",
    "Air_Pollution",
    "Alcohol_Use",
    "Smoking",
    "Obesity_Level",
    "Treatment_Cost_USD",
    "Survival_Years",
    "Target_Severity_Score",
]


categorical_columns = {
    "Gender",
    "Country_Region",
    "Cancer_Type",
    "Cancer_Stage",
}

target = "Cancer_type"
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Числовые признаки

for col_name in numeric_columns:
    plt.figure(figsize=(15, 5))
    sns.histplot(df[col_name], kde=True)
    plt.title(col_name)
    plt.show()

fig = plt.figure(figsize=(10, 7))
plt.boxplot(
    df[
        [
            "Genetic_Risk",
            "Air_Pollution",
            "Alcohol_Use",
            "Smoking",
            "Obesity_Level",
            "Survival_Years",
            "Target_Severity_Score",
        ]
    ]
)
plt.title("Ящик с усами")
plt.ylabel("Значения")
plt.xlabel("Данные")
plt.show()

fig = plt.figure(figsize=(10, 7))
plt.boxplot(df[["Age"]])
plt.title("Ящик с усами для стоимости")
plt.ylabel("Значения")
plt.xlabel("Данные")
plt.show()

fig = plt.figure(figsize=(15, 7))
plt.boxplot(df[["Treatment_Cost_USD"]])
plt.title("Ящик с усами для возраста")
plt.ylabel("Значения")
plt.xlabel("Данные")
plt.show()

# Категориальные признаки
# for col_name in categorical_columns:
#     plt.figure(figsize=(10, 5))
#     sns.countplot(x=df[col_name], order=df[col_name].value_counts().index)
#     plt.title(f"Распределение {col_name}")
#     plt.xticks(rotation=45)
#     plt.show()
```

```python
correlation = df[numeric_columns].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Корреляционная матрица")
plt.show()

# "Genetic_Risk"
# "Smoking"

# "Target_Severity_Score"


sns.scatterplot(x=df["Genetic_Risk"], y=df["Target_Severity_Score"])
plt.title("Genetic_Risk для Target_Severity_Score")
plt.show()

sns.scatterplot(x=df["Smoking"], y=df["Target_Severity_Score"])
plt.title("Smoking для Target_Severity_Score")
plt.show()
```
