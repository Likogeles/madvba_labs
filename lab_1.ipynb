```python
import matplotlib.pyplot as plt

def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.5)

def q3(x):
    return x.quantile(0.75)

def iqr(x):
    return q3(x) - q1(x)


df['Establishment Year'] = pd.to_numeric(df['Establishment Year'], errors='coerce')
df = df.dropna(subset=['Establishment Year'])
df['Establishment Year'] = df['Establishment Year'].astype(int)


df['Google review rating'] = pd.to_numeric(df['Google review rating'], errors='coerce')
df = df.dropna(subset=['Google review rating'])
df['Google review rating'] = df['Google review rating'].astype(float)


stats = df.groupby('Zone')['Establishment Year'].agg(['min', q1, 'median', q3, 'max', iqr])
print(stats)

df.boxplot(column='Establishment Year', by='Zone')
plt.show()

df['Establishment Year'].value_counts().sort_index().plot(kind='line', title='Распределение годов')
plt.show()

df['Zone'].value_counts().plot(kind='bar', title='Количество по типам')
plt.show()

df['Establishment Year'].plot(kind='hist', bins=30, title='Распределение годов')
plt.show()

df.plot(x='Establishment Year', y='Google review rating', kind='scatter')
plt.show()
```
